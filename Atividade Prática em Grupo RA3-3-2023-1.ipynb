```python
#Alunos: Joao Pedro Aires de Siqueira, Eduardo Pavellecini Belleboni
import threading as th
import time
import random

qnt_threads = 10
N = 10000

# Locks
lock = th.Lock()
resp_lock = [0] * qnt_threads

#-----------------------------Funçao wrapper -----------------------------------
def time_statistics(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Execution time of {func.__name__}: {execution_time} seconds\n\n")
        return result

    return wrapper


@time_statistics
def Lock():
    N = 10000
    qnt_threads = 10

    def carlao_lock(arg1, arg2, arg3, arg4, arg5):
        pass

    threads = [th.Thread(target=carlao_lock, args=(0, 1, N, resp_lock, i)) for i in range(qnt_threads)]
    for thread in threads:
        thread.start()
    for thread in threads:
        thread.join()
#-------------------------------------------------------------------------------

#samforos------------------
def sum_semaphore(semaphore, resp):
    for _ in range(qnt_threads):
        semaphore.release()
    print(f"Semaphore result: {sum(resp) / (N * qnt_threads)}")

def carlao_semaphore(start, end, N, resp, index, semaphore):
    count = 0
    for _ in range(N):
        x = random.random()
        y = random.random()

        if y < x**2:
            count += 1

    resp[index] = count
    semaphore.acquire()


@time_statistics
def semaphore_example():
    semaphore = th.Semaphore(qnt_threads)
    resp_semaphore = [0] * qnt_threads
    threads = [th.Thread(target=carlao_semaphore, args=(0, 1, N, resp_semaphore, i, semaphore)) for i in range(qnt_threads)] + [th.Thread(target=sum_semaphore, args=(semaphore, resp_semaphore))]
    for thread in threads:
        thread.start()
    for thread in threads:
        thread.join()


#locks------------------
@time_statistics
def lock_example():
    lock = th.Lock()
    resp_lock = [0] * qnt_threads
    threads = [th.Thread(target=carlao_lock, args=(0, 1, N, resp_lock, i)) for i in range(qnt_threads)]
    for thread in threads:
        thread.start()
    for thread in threads:
        thread.join()


#conditions-------------
condition = th.Condition()
resp_condition = [0] * qnt_threads

def carlao_condition(start, end, N, resp, index, condition):
    count = 0
    for _ in range(N):
        x = random.random()
        y = random.random()

        if y < x**2:
            count += 1

    with condition:
        resp[index] = count
        condition.notify()

def sum_condition(condition, resp):
    with condition:
        condition.wait_for(lambda: all(result != 0 for result in resp))
    print(f"Condition result: {sum(resp) / (N * qnt_threads)}")

@time_statistics
def condition_example():
    threads = [th.Thread(target=carlao_condition, args=(0, 1, N, resp_condition, i, condition)) for i in range(qnt_threads)] + [th.Thread(target=sum_condition, args=(condition, resp_condition))]
    for thread in threads:
        thread.start()
    for thread in threads:
        thread.join()


#barriers-------------------
barrier = th.Barrier(qnt_threads + 1)
resp_barrier = [0] * qnt_threads

def carlao_barrier(start, end, N, resp, index, barrier):
    count = 0
    for _ in range(N):
        x = random.random()
        y = random.random()

        if y < x**2:
            count += 1

    resp[index] = count
    barrier.wait()

def sum_barrier(barrier, resp):
    barrier.wait()
    print(f"Barrier result: {sum(resp) / (N * qnt_threads)}")

@time_statistics
def barrier_example():
    threads = [th.Thread(target=carlao_barrier, args=(0, 1, N, resp_barrier, i, barrier)) for i in range(qnt_threads)] + [th.Thread(target=sum_barrier, args=(barrier, resp_barrier))]
    for thread in threads:
        thread.start()
    for thread in threads:
        thread.join()


#Area de execuçao---------------------------------------------------------------
Lock()
lock_example()
semaphore_example()
condition_example()
barrier_example()
```
