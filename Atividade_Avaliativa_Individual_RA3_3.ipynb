# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

#definicao da funcao
valx = np.linspace(0, 2, 1000)
yalx = valx * np.exp(-valx**2)
yalx_resp = valx + 1
def f(p):
    x, y = p
    return [y - x * np.exp(-x**2), y - x - 1]

#definindo os limites das variaveis
resp_xval = []
for x1 in np.linspace(0, 1.5, 10):
    x_solution = fsolve(f, [x1, x1])
    resp_xval.append(x_solution[0])

resp_xval.append(2)


#definicao com a plotagem
calc = np.trapz(np.minimum(yalx, yalx_resp), valx)
plt.plot(valx, yalx_resp, label=r'$y = x + 1$', color='red')
plt.plot(valx, yalx, label=r'$y = xe^{-x^2}$', color='blue')
plt.fill_between(valx, np.minimum(yalx, yalx_resp), yalx_resp, where=(yalx <= yalx_resp), color='lightblue', alpha=0.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Area entre as curvas', fontsize=14)
plt.legend()

plt.show()
```
